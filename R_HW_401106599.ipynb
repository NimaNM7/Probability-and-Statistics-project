    تابعی بنویسید تا با دریافت طول ویدئو به عنوان ورودی، احتمال تماشای بدون وقفه ویدئو را تخمین بزند. به منظور تخمین احتمال، می‌توانید تماشای ویدئو را به تعداد ۱۰۰۰۰ بار شبیه‌سازی کرده و سپس نسبت تعداد تجربیات خوشایند به کل آزمایشات را به جای احتمال در نظر بگیرید.


```r
# Do not change this cell
# playback rate
rate <- 1
# lower bound of download rate
lbound <- 0.6
# upper bound of download rate
ubound <- 1.4
# number of simulations
n_sim = 10000
```

```r
check_pleasant1 <- function(length) {
  a <- runif(length, lbound, ubound)
  cumulative_sums <- cumsum(a)
  checks <- cumulative_sums >= 1:length
  return(all(checks))
}

bufferrate1 <- function(length) {
  sample_results <- sapply(1:n_sim, function(x) check_pleasant1(length))
  number_of_success <- sum(sample_results)
  return (number_of_success / n_sim)
}
```

 طول ویدئو را همواره ۱۰۰ ثانیه در نظر بگیرید.  نمودار احتمال تماشای بدون وقفه را نسبت به مقادیر مختلف برای زمان headstart رسم کنید. می‌توانید مقادیر زمان headstart را بین ۰ تا ۴۰ ثانیه با گام‌های ۱ ثانیه پوشش دهید.

```r
# Do not change this cell
# playback rate
rate <- 1
# lower bound of download rate
lbound <- 0.2
# upper bound of download rate
ubound <- 1.5
# number of simulations
n_sim = 10000
```

```r
check_pleasant2 <- function(length, headstart) {
  a <- runif(length, 0.2, 1.5)
  cumulative_sums <- cumsum(a)
  checks <- cumulative_sums >= (1 - headstart):(length-headstart)
  return(all(checks))
}
# modify the bufferrate function and include the headstart
bufferrate2 <- function(length, headstart) {
  sample_results <- sapply(1:n_sim, function(x) check_pleasant2(length, headstart))
  number_of_success <- sum(sample_results)
  return (number_of_success / n_sim)
}
```

```r
# define headstarts
# calculate the probability against headstart values
probs <- sapply(0:40, function(x) bufferrate2(100,x))

# plot the probabilities
df <- data.frame(x = (0:40), y = probs)

ggplot(df, aes(x = x, y = y)) + 
  geom_point() + 
  labs(title = "Plot of Probability based of Headstart", x = "Headstart", y = "Probability")
```

```r
# Do not change this cell in your final notebook
t1 <- 30
t2 <- 90
cookTime <- 360
n_pieces <- 20
sides <- 6
```

```r
# check if potatoes are coocked properly
# cookTime is the matrix of amount of time that each side of each potatoe is cooked
# underCookTime and burnTime are the times between which the amount of cooking should be in order to be proper
# return T if all the cooktimes are between the two values burnTime and cookTime otherwise return F
cookTest <- function(underCookTime, burnTime, cookTime) {
  rows <- nrow(cookTime)
  cols <- ncol(cookTime)

  result <- all(cookTime > underCookTime & cookTime < burnTime)
  result
}
```

```r
# simulate one shuffling and return the new times
# cookTime is the cooked times before the current time simulation
# n_pieces is the number of pieces used in simulation
# sectionTime is the amount of time each shuffling lasts
# sides is the number of sides each potatoe has
sectionSimulate <- function(cookTime, n_pieces, sectionTime, sides) {
  cookedSides <- floor(runif(n_pieces, 1, sides + 1))
  cookTime[cbind(1:n_pieces, cookedSides)] <- cookTime[cbind(1:n_pieces, cookedSides)] + sectionTime
  cookTime
}
```

```r
# timeLength is the total time of simulation
# n_shuffle is the number of times potatoes are changed sides
# implement a complete simulation here and return if this was a successful cooking using the cookTest function
cookSimulation <- function(underCookTime,burnTime,n_pieces,timeLength,sides,n_shuffle){
  cookMatrix <- matrix(0, nrow = n_pieces, ncol = sides)
  section_length <- timeLength / n_shuffle
  for (i in 1:n_shuffle) {
    cookMatrix <- sectionSimulate(cookMatrix, n_pieces, section_length, sides)
  }
  answer <- cookTest(underCookTime, burnTime, cookMatrix)
  answer
}
```

```r
# the function that calculates probability of success - it tests with 500 samples
# this function can take too long so be patient please
probability <- function(underCookTime,burnTime,n_pieces,timeLength,sides,n_shuffle) {
  n <- 400
  samples <- sapply(1:n, function(x) cookSimulation(underCookTime,burnTime,n_pieces,timeLength,sides,n_shuffle))
  trues <- sum(samples)
  trues/ n
}
```

```r
# do the simulation for values of shuffling number between 80 and 400 with a stride of 5
# sampling from 80 to 400 shuffles and putting the result in samples
x = seq(80, 400, by = 5)
samples <- sapply(x, function(x) probability(t1, t2, n_pieces, cookTime, sides, x))
```
